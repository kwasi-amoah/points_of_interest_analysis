```python
import pandas as pd
from geopy.distance import geodesic

# Load the Excel sheets
pois = pd.read_excel("File path")
branches = pd.read_excel("File path")
atms = pd.read_excel("File path")

# Initialize a list to store results
results = []

# Loop through each POI
for _, poi in pois.iterrows():
    poi_coords = (poi["Latitude"], poi["Longitude"])

    # Find the closest bank branch
    closest_branch = None
    closest_branch_distance = float('inf')
    for _, branch in branches.iterrows():
        branch_coords = (branch["Latitude"], branch["Longitude"])
        distance = geodesic(poi_coords, branch_coords).kilometers
        if distance < closest_branch_distance:
            closest_branch_distance = distance
            closest_branch = branch

    # Find the closest ATM
    closest_atm = None
    closest_atm_distance = float('inf')
    for _, atm in atms.iterrows():
        atm_coords = (atm["Latitude"], atm["Longitude"])
        distance = geodesic(poi_coords, atm_coords).kilometers
        if distance < closest_atm_distance:
            closest_atm_distance = distance
            closest_atm = atm

    # Append the results for the current POI
    results.append({
        "Name": poi["Name"],
        "Amenity": poi["Amenity"],
        "POI Latitude": poi["Latitude"],
        "POI Longitude": poi["Longitude"],
        "Closest Bank Branch": closest_branch["Name"],
        "Distance to Branch (km)": closest_branch_distance,
        "Closest ATM": closest_atm["Name"],
        "Distance to ATM (km)": closest_atm_distance
    })

# Create a DataFrame with the results
results_df = pd.DataFrame(results)

results_df
```

```python
# Save the results to a new Excel file
output_path = "FIle name"  # Specify your desired file name
results_df.to_excel(output_path, index=False, engine="openpyxl")
print(f"Results saved to {output_path}")
```
